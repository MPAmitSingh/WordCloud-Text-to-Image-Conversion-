```python
# Create A list 
words=('Python Python Python Matplotlib Matplotlib Seaborn Network Plot Violin Chart Pandas Datascience Wordcloud Spider Radar Parrallel Alpha Color Brewer Density Scatter Barplot Barplot Boxplot Violinplot Treemap Stacked Area Chart Chart Visualization Dataviz Donut Pie Time-Series Wordcloud Wordcloud Sankey Bubble')
```

```python
words
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Create the wordcloud object
wordcloud=WordCloud(width=500,height=500,margin=0).generate(words)
```

```python
# Display the generate image.
plt.imshow(wordcloud,interpolation='bicubic')
plt.axis('off')
plt.margins(x=0,y=0)
plt.show()
```
